### Difference Using rank_by as: "Distance" or "Prominence"

Let's see how the venues are scattered around our search coordinate if we change the rank_by parameter.

For both examples we are going to use the coordinates saved in "neighborhoods.csv"

```python
import pandas as pd
df=pd.read_csv('dataset/shape_neighborhoods_df.csv', index_col=0)
```

```python
df.head()
```

```python
def extract_values(obj, key):
    """Pull all values of specified key from nested JSON."""
    arr = []

    def extract(obj, arr, key):
        """Recursively search for values of key in JSON tree."""
        if isinstance(obj, dict):
            for k, v in obj.items():
                if isinstance(v, (dict, list)):
                    extract(v, arr, key)
                elif k == key:
                    arr.append(v)
        elif isinstance(obj, list):
            for item in obj:
                extract(item, arr, key)
        return arr

    results = extract(obj, arr, key)
    return results

def geometry_cleaner(df):
    """Clean a geometry column with lat and lng"""
    latitude=[]
    longitude=[]
    for i in df.index:
        text=df.loc[i,"geometry"]
        lat=extract_values(text, 'lat')
        long=extract_values(text, 'lng')
        latitude.append(lat[1])
        longitude.append(long[1])
    df['Latitude']=latitude
    df['Longitude']=longitude
    df.drop(columns=['geometry'],inplace=True)
```
