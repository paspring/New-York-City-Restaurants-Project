```python
import requests
import geopandas as gpd
from shapely.geometry import shape

data="dataset/nta.geojson"

shapes=geopandas.read_file(data)
```

```python
shapes.head()
```

```python
from shapely.geometry import Point, Polygon
```

```python
import pandas as pd
restaurants=pd.read_csv('dataset/foursquare_with_rating_537.csv',index_col=0)
restaurants=restaurants.reset_index(drop=True)
restaurants['Borough']=""
restaurants['Neighborhood']=""
restaurants['GeoID']=""
```

```python
restaurants.head()
```

```python
import time
for i in restaurants.index:
    lng=restaurants.lat[i]
    lat=restaurants.lng[i]

    for j in shapes.index:
        start_time = time.time()
        pol=shapes.geometry[j]
        point=Point(lat,lng)
        probe=Point(lng,lat).within(pol)
        cent=pol.centroid
        probe=pol.contains(point)
        probe2=pol.contains(cent)
        if probe == True:
            Neighborhood_Restaurant=shapes.ntaname[j]
            Borough_Restaurant=shapes.boro_name[j]
            GeoID=shapes.ntacode[j]
            #print(i,restaurants_nyc_final.name[i],Borough_Restaurant,Neighborhood_Restaurant)
            restaurants.loc[i,'Borough']=Borough_Restaurant
            restaurants.loc[i,'Neighborhood']=Neighborhood_Restaurant
            restaurants.loc[i,'GeoID']=GeoID
            
            break
        else:
            continue
    print(i)
    print("--- %s seconds ---" % (time.time() - start_time))
```
