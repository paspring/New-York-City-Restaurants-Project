```python
import pandas as pd 
foursquare=pd.read_csv('dataset/restaurants_only.csv',index_col=0)
print(foursquare.shape)
foursquare.head()
```

```python
c=foursquare.Neighborhood.value_counts()
b=foursquare.Borough.value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(c.index[0:10],c[0:10])
ax.set_xticklabels(c.index, rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.bar(b.index,b)
ax.set_xticklabels(b.index, rotation=90)
plt.show()
```
